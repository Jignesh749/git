```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
from sklearn.cluster import KMeans
import pandas as pd

# Set up Spotify API credentials
client_id = "716c1e25d0b94ad59424c2fe6e5268ec"
client_secret = "1f967480693941c69c6265ca6d920b4f"
redirect_uri = "http://localhost:8001/callback"
scope = "playlist-modify-private user-top-read"

# Initialize the Spotify client
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=client_id, client_secret=client_secret, redirect_uri=redirect_uri, scope=scope))

def cluster_tracks(track_ids, num_clusters=3):
    # Initialize a list to store features and track names
    features = []
    track_names = []

    # Fetch audio features and track details for each track ID
    for track_id in track_ids:
        # Fetch audio features
        audio_features = sp.audio_features([track_id])[0]
        
        # Check if audio features are available
        if audio_features is None:
            continue
        
        # Fetch track details (name)
        track_details = sp.track(track_id)
        track_name = track_details['name']
        track_names.append(track_name)

        # Append audio features to features list
        features.append([
            audio_features.get('danceability', None),
            audio_features.get('energy', None),
            audio_features.get('valence', None),
            audio_features.get('tempo', None),
            audio_features.get('speechiness', None),
            audio_features.get('acousticness', None),
            audio_features.get('instrumentalness', None),
            audio_features.get('liveness', None),
            audio_features.get('loudness', None)
        ])

    # Convert features to a DataFrame
    if features:
        feature_df = pd.DataFrame(features, columns=[
            'danceability', 'energy', 'valence', 'tempo',
            'speechiness', 'acousticness', 'instrumentalness',
            'liveness', 'loudness'
        ])

        # Cluster tracks using K-Means
        kmeans = KMeans(n_clusters=num_clusters, random_state=42)
        feature_df['cluster'] = kmeans.fit_predict(feature_df)

        # Add track names to the DataFrame
        feature_df['track_name'] = track_names

        # Print the DataFrame with clusters and track names
        print(feature_df)

        # Return DataFrame with clusters
        return feature_df
    else:
        return None

# Example usage: cluster tracks
track_ids = ['72vuBPMhwFNlSYpTSf6fVD', '4eBvRhTJ2AcxCsbfTUjoRp', '5ZLkihi6DVsHwDL3B8ym1t', '6wkHR8cU4INbp145hngbQO', '6c4Po8zyQjB4cR5XgLhcGb', '4BDOA9pV0yLClbR3v37ekI', '3Jd1yxcUQLdfBJcsxFtYzj', '610keNiNVTIkYD0CqeSerg', '4K5aBuakMjnDhFa56CmlHC', '209QxKVJzh2FLUmZHpYs4m']
clustered_tracks = cluster_tracks(track_ids, num_clusters=3)

if clustered_tracks is not None:
    print(clustered_tracks)
else:
    print("No clustering results available due to invalid or missing audio features.")
```
